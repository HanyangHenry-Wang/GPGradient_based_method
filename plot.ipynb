```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def log_RegretMean_plot(res,fstar,label,color,linestyle,error_bound=True):
    
    
    total_record = (fstar-np.array(res))
    
    
    N = total_record.shape[0]
    
    std = np.std(total_record,axis=0)
    
    mean = np.mean(total_record,axis=0)
    
    median = np.median(total_record,axis=0)
    
    iter = np.array(range(mean.shape[0]))
    
    lower = mean - 1*std/np.sqrt(N)
    upper = mean + 1*std/np.sqrt(N)
    
    lower.astype(float)
    upper.astype(float)
    
    
    plt.plot(mean,linewidth=2,label=label,color=color,linestyle=linestyle)
    if error_bound:
        plt.fill_between(iter, lower, upper,alpha=0.25,color=color)   
    
```

```python
def log_Regret_plot(res,fstar,label,color,linestyle,error_bound=True):
    
    
    total_record = np.log10(fstar-np.array(res))
    
    
    N = total_record.shape[0]
    
    std = np.std(total_record,axis=0)
    
    mean = np.mean(total_record,axis=0)
    
    median = np.median(total_record,axis=0)
    
    iter = np.array(range(mean.shape[0]))
    
    lower = mean - 1*std/np.sqrt(N)
    upper = mean + 1*std/np.sqrt(N)
    
    lower.astype(float)
    upper.astype(float)
    
    
    plt.plot(mean,linewidth=2,label=label,color=color,linestyle=linestyle)
    if error_bound:
        plt.fill_between(iter, lower, upper,alpha=0.25,color=color)   
```

```python
EI = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Ackley2D_ExpectedImprovement'), delimiter=',')
# TEI = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Beale2D_TruncatedExpectedImprovement'), delimiter=',')
# MES = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Beale2D_MES_KnownOptimum'), delimiter=',')
PDF = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Ackley2D_Fstar_pdf'), delimiter=',')
#PDF_gradient = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Branin2D_Fstar_pdf_GradientEnhanced'), delimiter=',')
PDF_gradient_f = np.genfromtxt(os.path.join(os.getcwd(),'exp_res/Ackley2D_Fstar_pdf_GradientEnhanced_fantasy'), delimiter=',')

fstar = 0.

log_RegretMean_plot(EI,fstar=fstar,label='EI',color='grey',linestyle='dotted',error_bound=True)
# log_RegretMean_plot(TEI,fstar=fstar,label='TEI',color='blue',linestyle='dashed',error_bound=True)
# log_RegretMean_plot(MES,fstar=fstar,label='MES',color='darkviolet',linestyle='-',error_bound=True)
log_RegretMean_plot(PDF,fstar=fstar,label='PDF',color='yellow',linestyle='-',error_bound=True)
#log_RegretMean_plot(PDF_gradient,fstar=fstar,label='PDF+gradient',color='black',linestyle='-',error_bound=True)
log_RegretMean_plot(PDF_gradient_f,fstar=fstar,label='PDF+gradient_fantasy',color='black',linestyle='-',error_bound=True)

plt.xlabel('iteration')
plt.ylabel('regret')

plt.grid()

plt.yscale("log") 
plt.grid()
plt.legend()
plt.title('Ackley 2D')
```
